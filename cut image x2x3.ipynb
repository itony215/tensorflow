```python
import csv
import os
import random
import sys
from datetime import datetime
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
import SimpleITK as sitk
from time import strftime
mass_training_file = "D:\\DDSM\\total_case_path_train_density.csv"
f = open(mass_training_file, 'r')
def rotate90(src):
        img_copy = src.copy()
        img_copy = cv2.transpose(img_copy)
        img_copy = cv2.flip(img_copy, 1)
        return img_copy
def flip(src):
        img_copy = src.copy()
        img_copy = cv2.flip(img_copy, 1)
        return img_copy    
def read_dcm_image(path):
        ds = sitk.ReadImage(path)
        img_array = sitk.GetArrayFromImage(ds)
        return img_array[0]
a=1
b=1
c=1
d=1    
num_line=0
for line in f:
    if num_line >=1:
        break
    num_line += 1
    line = line.rstrip()
    l = line.split(",")
    if len(l) <= 2:
        continue
    print("Loading training data:" + l[0])
    image_path = l[2] + l[3]
    img = (read_dcm_image(image_path))
    
    print(img.shape)
    
    howmuch=2
    for i in range(howmuch):
        splity=[]
        splitx=[]
        howbig= random.randint(1000,3000)
        print('howbig',howbig)
        for i in range(img.shape[1]):
            if( sum( img[:,i] )/img.shape[0] > howbig ):
                splity.append(i)
        #print(splity)
        for i in range(img.shape[0]):
            if( sum( img[i,:] )/img.shape[1] > howbig ):
                splitx.append(i)
        #print(splitx)
        count=0
        start=[]
        find=[]
        splitx.append(0)
        for i in range(len(splitx)-1):
            if splitx[i+1]-splitx[i]==1:
                if count==0:
                    start.append(i)
                count=count+1
            else:
                if count != 0:
                    find.append(count)
                    #print('x_len',count)
                count=0
        p=find.index(max(find)) 

        county=0
        starty=[]
        findy=[]
        splity.append(0)
        for i in range(len(splity)-1):
            if splity[i+1]-splity[i]==1:
                if county==0:
                    starty.append(i)
                county=county+1
            else:
                if county != 0:
                    findy.append(county)
                    #print('y_len',county)
                county=0
        py=findy.index(max(findy)) 
        print('x_start',start)
        print('x_length',find)
        print('y_start',starty)
        print('y_length',findy)
        print('splitx[start[p]]',splitx[start[p]])
        print('splitx[start[p]+find[p]]',splitx[start[p]+find[p]])
        print('splity[starty[py]]',splity[starty[py]])
        print('splity[starty[py]+findy[py]]',splity[starty[py]+findy[py]])
        cutimg=img[splitx[start[p]]:splitx[start[p]+find[p]],splity[starty[py]]:splity[starty[py]+findy[py]]]
        '''
        plt.imshow(cutimg ,cmap='gray')
        plt.title('cut background image')
        plt.axis('off')
        plt.show()
        '''
        smallimage=cv2.resize(cutimg,(224,224), interpolation=cv2.INTER_CUBIC)
        d1=rotate90(smallimage)
        d2=rotate90(d1)
        d3=rotate90(d2)
        f1=flip(smallimage)
        f2=flip(d1)
        f3=flip(d2)
        f4=flip(d3)
        print(l[1])
        if l[1]==str(1):
            cv2.imwrite('C'+l[1]+'_'+str(a)+'.png', smallimage)
            a=a+1
            cv2.imwrite('C'+l[1]+'_'+str(a)+'.png', d1)
            a=a+1
            cv2.imwrite('C'+l[1]+'_'+str(a)+'.png', d2)
            a=a+1
            cv2.imwrite('C'+l[1]+'_'+str(a)+'.png', d3)
            a=a+1
            cv2.imwrite('C'+l[1]+'_'+str(a)+'.png', f1)
            a=a+1
            cv2.imwrite('C'+l[1]+'_'+str(a)+'.png', f2)
            a=a+1
            cv2.imwrite('C'+l[1]+'_'+str(a)+'.png', f3)
            a=a+1
            cv2.imwrite('C'+l[1]+'_'+str(a)+'.png', f4)
            a=a+1
            print("save1")
        elif l[1]==str(2):
            cv2.imwrite('C'+l[1]+'_'+str(b)+'.png', smallimage)
            b=b+1
            cv2.imwrite('C'+l[1]+'_'+str(b)+'.png', d1)
            b=b+1
            cv2.imwrite('C'+l[1]+'_'+str(b)+'.png', d2)
            b=b+1
            cv2.imwrite('C'+l[1]+'_'+str(b)+'.png', d3)
            b=b+1
            cv2.imwrite('C'+l[1]+'_'+str(b)+'.png', f1)
            b=b+1
            cv2.imwrite('C'+l[1]+'_'+str(b)+'.png', f2)
            b=b+1
            cv2.imwrite('C'+l[1]+'_'+str(b)+'.png', f3)
            b=b+1
            cv2.imwrite('C'+l[1]+'_'+str(b)+'.png', f4)
            b=b+1
            print("save2")
        elif l[1]==str(3):
            cv2.imwrite('C'+l[1]+'_'+str(c)+'.png', smallimage)
            c=c+1
            cv2.imwrite('C'+l[1]+'_'+str(c)+'.png', d1)
            c=c+1
            cv2.imwrite('C'+l[1]+'_'+str(c)+'.png', d2)
            c=c+1
            cv2.imwrite('C'+l[1]+'_'+str(c)+'.png', d3)
            c=c+1
            cv2.imwrite('C'+l[1]+'_'+str(c)+'.png', f1)
            c=c+1
            cv2.imwrite('C'+l[1]+'_'+str(c)+'.png', f2)
            c=c+1
            cv2.imwrite('C'+l[1]+'_'+str(c)+'.png', f3)
            c=c+1
            cv2.imwrite('C'+l[1]+'_'+str(c)+'.png', f4)
            c=c+1
            print("save3")
        elif l[1]==str(4):
            cv2.imwrite('C'+l[1]+'_'+str(d)+'.png', smallimage)
            d=d+1
            cv2.imwrite('C'+l[1]+'_'+str(d)+'.png', d1)
            d=d+1
            cv2.imwrite('C'+l[1]+'_'+str(d)+'.png', d2)
            d=d+1
            cv2.imwrite('C'+l[1]+'_'+str(d)+'.png', d3)
            d=d+1
            cv2.imwrite('C'+l[1]+'_'+str(d)+'.png', f1)
            d=d+1
            cv2.imwrite('C'+l[1]+'_'+str(d)+'.png', f2)
            d=d+1
            cv2.imwrite('C'+l[1]+'_'+str(d)+'.png', f3)
            d=d+1
            cv2.imwrite('C'+l[1]+'_'+str(d)+'.png', f4)
            d=d+1
            print("save4")
        else:
            print("XXX")
```
